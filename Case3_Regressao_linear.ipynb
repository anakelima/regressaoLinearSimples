```python
## variaveis explicativa(X) e preditivas(Y)
x = base_casa[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_Rooms',
       'Avg_Area_Number_Bedrooms', 'Area_Population']]
y = base_casa[['Price']]
```

```python
## treinando meu machine learning
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.7, test_size = 0.3, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
y_pred = lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r = r2_score(y_test,y_pred)
print("r_quadrado", r)
```

```python
## pegando a biblioteca para plotar
import matplotlib.pyplot as plt

c = [i for i in range(1,1501,1)]
fig = plt.figure(figsize = (12,8))
plt.plot(c, y_test, color = "blue") ## base real
plt.plot(c, y_pred, color = "red") ##predição de teste
plt.xlabel("index")
plt.ylabel("Preço do Imóvel")
```
